## 1. Matplotlib plots <a name="plots"></a> 

What is a plot in Matplotlib? A plot is a graphical representation of your data. Matplotlib plots your data, in Jupyter called widgets (html-based windows). A plot can be defined by one or more axes. 

In this tutorial, we will use a collection of command-style functions. Matplotlib.pyplot allows us to work with Matplotlib similar to the graphical tools in MATLAB.

Let's import the Matplotlib library and create a simple plot.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,10,100) #returns an array of a hundred values between 0 and 10, with same distance
y = np.exp(x)
plt.plot(x,y)
plt.show()
```

According to the matplotlib documentation, you can also change the color and line style of the plot by using the third argument of each x,y pair. 

You can also plot more than one graph by adding more arguments to the plot function.

```python
x = np.arange(0., 5., 0.2)
y = np.exp(x)
z = x**2
plt.plot(x,y,'ro', x,z,'g^')
plt.show()
```

To make your plot more compelling and easier to understand, you can use matplotlib's `axis labeling`, `title` and `legend` functions. 

```python
x = np.arange(0., 5., 0.2)
y = np.exp(x)
z = x**2
plt.plot(x,y,'ro', label ="exponential")
plt.plot(x,z,'g^', label ="quadratic")
plt.xlabel('Time')
plt.ylabel('Individuals')
plt.title('Exponential vs. quadratic growth')
plt.legend()

plt.show()
```

You can also customize your axes. The `axes' of a matplotlib plot are number-line objects and set the boundaries of the graph, generating the `ticks' and `ticklabels'.

```python
x = np.arange(0., 5., 0.2)
y = np.exp(x)
z = x**2
plt.plot(x,y,'ro', label ="exponential")
plt.plot(x,z,'g^', label ="quadratic")

plt.axis([0, 4, 0, 50])
plt.grid(True)
plt.xticks([0,1,2,3])
plt.xticks([0,1,2,3,4], labels = ['zero','one','two','three','four'])

plt.xlabel('Time')
plt.ylabel('Individuals')
plt.title('Exponential vs. quadratic growth')
plt.legend()
plt.show()
```

## 2. Matplotlib types of plots <a name="types"></a> 

Matplotlib includes several types of plots for all kinds of applications. It also allows to create subplots to create more than one plot in a single widget. In this example, we will however still use the line plot.

```python
x = np.arange(0., 5., 0.2)
y = np.exp(x)
z = x**2

plt.subplot(2, 1, 1)
plt.plot(x,y,'ro', label ="exponential")
plt.ylabel('Individuals')
plt.title('Exponential vs. quadratic growth')

plt.subplot(2, 1, 2)
plt.plot(x,z,'g^', label ="quadratic")
plt.xlabel('Time')
plt.ylabel('Individuals')
plt.show()
```

Next, we consider scatterplots. Scatterplots are generally useful for identifying groups and structures in the data.

```python
x = np.random.rand(20)
y = np.random.rand(20)
plt.scatter(x, y,  c=np.random.rand(20), alpha=0.8)
plt.show()
```

You can also use histograms to show the distribution of your data. Let's remember the interest rate example from the numpy introduction. The `bins` property allows you to specify the number of intervals in the histogram.

```python
r1 = np.array([0.05, 0.01, 0.1, 0.12, 0.06, 0.04, 0.05])
print(np.histogram(r1))

plt.hist(r1,bins=10)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogramm')
plt.show()
```
